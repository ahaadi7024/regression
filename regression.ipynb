1. What is Simple Linear Regression?

Simple Linear Regression is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y).

Equation:

Y=mX+c

Where:

Y = Dependent variable
X = Independent variable
m = Slope
c = Intercept
2. What are the key assumptions of Simple Linear Regression?
Linear relationship between X and Y
Independence of observations
Homoscedasticity (constant variance of errors)
Normally distributed residuals
No significant outliers
3. What does the coefficient m represent in Y = mX + c?

The coefficient m is called the slope.

It shows how much Y changes when X increases by one unit.

Example:
If

Y=2X+5

then for every 1-unit increase in X, Y increases by 2 units.

4. What does the intercept c represent in Y = mX + c?

The intercept c is the value of Y when X = 0.

Example:

Y=2X+5

When X = 0,

Y=5

10. Key assumptions of Multiple Linear Regression
Linear relationship
Independence of errors
Homoscedasticity
Normal distribution of residuals
No multicollinearity
No significant outliers
11. What is Heteroscedasticity?

Heteroscedasticity occurs when the variance of residuals is not constant.

Effect:
Unreliable coefficient estimates
Incorrect p-values
Poor predictions
12. How can multicollinearity be reduced?
Remove highly correlated features
Use PCA
Use Ridge Regression
Collect more data

15. How does intercept interpretation differ?
Simple Regression

Intercept = Y when X = 0.

Multiple Regression

Intercept = Y when all independent variables are 0.

16. What is the significance of slope?

The slope shows how much Y changes for a one-unit increase in a predictor variable.

Positive slope → Positive relationship

Negative slope → Negative relationship

17. How does intercept provide context?

The intercept acts as a baseline prediction when predictor variables are zero.

It helps understand where the regression line starts.

18. Limitations of R²
Does not indicate causation
Can increase with unnecessary variables
Does not detect overfitting
Doesn't guarantee a good model
19. What does a large standard error indicate?

A large standard error means:

Coefficient estimates are less reliable
High variability exists
Predictor may not be significant
20. How identify heteroscedasticity in residual plots?

Plot residuals vs predicted values.

Signs:
Funnel shape
Increasing spread
Decreasing spread
Why important?

Violates regression assumptions and affects statistical tests.

21. High R² but Low Adjusted R² means?

It indicates:

Too many unnecessary features
Possible overfitting
Some predictors contribute little information
22. Why scale variables in Multiple Linear Regression?

Scaling helps:

Faster convergence

Fair comparison of coefficients
Improves regularized models (Ridge, Lasso)

31. How is Polynomial Regression implemented in Python?
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 4, 9, 16, 25])

# Convert to polynomial features
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train model
model = LinearRegression()
model.fit(X_poly, y)

# Prediction
y_pred = model.predict(X_poly)

print(y_pred)

This code creates polynomial features, trains a regression model, and makes predictions.
